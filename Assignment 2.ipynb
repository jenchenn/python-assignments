```python
#center and scale "age" and "bmi" (normal distributions) to view distribution on a smaller scale
df_norm = df[['age','bmi']]
df_norm.head()
```

```python
df_norm.hist()
```

```python
from sklearn import preprocessing as prep
```

```python
scaled_data = prep.scale(df_norm)
scaled_insurance_df = pd.DataFrame(scaled_data, 
                                   index=df_norm.index,
                                   columns=df_norm.columns)
scaled_insurance_df.describe()
```

```python
#double check scaled distributions
scaled_insurance_df.hist()
```

```python
#transform "charges" (right skewed) b/c difference is too big
df_skew = df[['charges']]
df_skew.head()
```

```python
df_skew.hist()
```

```python
#apply square root transformation
sqrt_transformed_col = df_skew.apply(np.sqrt) #sqrt of data points
sqrt_transformed_col.hist() #plot histogram to check
```

```python
#try with log transformation
log_transformed_col = (df_skew+1).apply(np.log) #log of data points
log_transformed_col.hist() #plot histogram to check

#log transformation is better represented in this case, since it is closer to a normal distribution!
```

```python
#now to check for missing string variables in sex, smoker, region
#and fill in missing data with whatever occurs most frequently

#examine "sex" column
df["sex"].isnull().sum()
#3 missing values
#after filling missing data = 0
```

```python
df.groupby(["sex"]).agg({'sex':'count'})
```

```python
#"male" occurs more frequently
df["sex"].fillna("male", inplace=True)
```

```python
#check data type of "sex" column values
df["sex"].dtype
```

```python
#examine "smoker" column
df["smoker"].isnull().sum()
#no missing data
```

```python
#check data type of "smoker" column values
df["smoker"].dtype
```

```python
#last but not least, examine "region" column
df["region"].isnull().sum()
#5 missing values
#after filling missing data = 0
```

```python
df.groupby(["region"]).agg({'region':'count'})
```

```python
#"southeast" occurs most frequently
df["region"].fillna("southeast", inplace=True)
```
